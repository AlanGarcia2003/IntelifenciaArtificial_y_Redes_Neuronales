```python
#1. Cargar librerías
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
#2. Cargar datos
np.random.seed(0)
horas_estudio = np.random.rand(100, 1) * 10
calificaciones = 2 * horas_estudio + np.random.randn(100,1) * 2
```

```python
#3. Preprocesamiento de datos
train_data, test_data, train_labels, test_labels = train_test_split(horas_estudio, calificaciones, test_size=0.2, random_state=42)
```

```python
#4. Selección y entrenamiento del modelo
modelo = LinearRegression()
modelo.fit(train_data, train_labels)
```

```python
#5. Prueba de modelo
predicciones = modelo.predict(test_data)
mse = mean_squared_error(test_labels, predicciones)
print("ERROR:", mse)
```

```python
# RESULTADOS
plt.scatter(train_data, train_labels, color='red', label='Datos de entrenamiento')
plt.scatter(test_data, test_labels, color='black', label='Datos de prueba')
plt.plot(test_data, predicciones, color='blue', label='Modelo de regresión lineal')
plt.xlabel('Horas de estudio')
plt.ylabel('Calificación')
plt.title('Modelo de regresión lineal para predicción de calificaciones')
plt.legend()
plt.show()
```
